### Libraries

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Data set of 10 images

```python
# Define image parameters
img_height, img_width = 128, 128  
batch_size = 32

data_dir = '/Users/milind/Documents/try/Untitled Folder/Data'  

# Data augmentation and preprocessing
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)
```

## ANN Model with 3 hidden layes and LeakyReLU
* 100% accuracy

```python
from tensorflow.keras.layers import LeakyReLU

# Define image parameters
img_height, img_width = 128, 128  
batch_size = 32

# Build the ANN model
model = Sequential([
    Flatten(input_shape=(128, 128, 3)),
    Dense(128, activation=LeakyReLU(alpha=0.01)),
    Dense(64, activation=LeakyReLU(alpha=0.01)),
    Dense(64, activation=LeakyReLU(alpha=0.01)),    
    Dense(1, activation='sigmoid')  # Binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(
    train_generator,
    epochs=50,
    validation_data=validation_generator
)

# Evaluate the model
loss, accuracy = model.evaluate(validation_generator)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```

```python
from kerastuner.tuners import RandomSearch

def build_model(hp):
    model = Sequential([
        Flatten(input_shape=(128, 128, 3)),
        Dense(hp.Int('units', min_value=64, max_value=256, step=64), activation=LeakyReLU(alpha=0.01)),
        Dense(64, activation=LeakyReLU(alpha=0.01)),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=hp.Choice('optimizer', ['adam', 'sgd']),
                  loss='binary_crossentropy', metrics=['accuracy'])
    return model

tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    directory='tuner_results'
)

tuner.search(train_generator, epochs=10, validation_data=validation_generator)
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"Best Parameters: {best_hps.values}")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define image parameters
img_height, img_width = 128, 128  
batch_size = 32

data_dir = '/Users/milind/Documents/try/Untitled Folder/Data'

# Apply Data Augmentation
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)

# Prepare augmented datasets
train_generator_100 = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)

validation_generator_100 = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)
```

```python
# Build the ANN model
model = Sequential([
    Flatten(input_shape=(128, 128, 3)),
    Dense(128, activation=LeakyReLU(alpha=0.01)),
    Dense(128, activation=LeakyReLU(alpha=0.01)),
    Dense(64, activation=LeakyReLU(alpha=0.01)),
    Dense(64, activation=LeakyReLU(alpha=0.01)),    
    Dense(1, activation='sigmoid')  # Binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(
    train_generator_100,
    epochs=100,
    validation_data=validation_generator_100
)

# Evaluate the model
loss, accuracy = model.evaluate(validation_generator_100)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```
